```python
import numpy as np
import matplotlib.pyplot as plt
def gaussian_function(x, y, *args):
    
    sigma_x, sigma_y, mu_x, mu_y = args
    x_now = x - mu_x
    y_now = y - mu_y
    return np.exp(-(x_now*x_now)/(2*sigma_x**2) - (y_now*y_now)/(2*sigma_y**2))/( np.sqrt(2 * np.pi) * sigma_x * sigma_y)



def image_making(x_len, y_len, x_lim, y_lim, function = gaussian_function, args = (5, 5, 100, 100)):
    xmin, xmax = x_lim
    ymin, ymax = y_lim
    x = np.linspace(xmin, xmax, x_len)
    y = np.linspace(ymin, ymax, y_len)
    xx, yy= np.meshgrid(x, y, indexing='ij')
    return function(xx, yy, *args), xx, yy



x_len, y_len = 301, 301
images, xx, yy = image_making(x_len, y_len, (-1,1),(-1,1), gaussian_function, (0.01,0.01, 0,0))
print(np.max(images))
plt.imshow(xx.T, origin = "lower")
plt.show()
plt.imshow(yy.T, origin = "lower")
plt.show()
plt.imshow(images.T, origin = "lower")
plt.show()
```

```python
plot_ = True
## FFT
fft_now = np.fft.fft2(images)

## Shifting origin 
fft_now = np.fft.fftshift(fft_now)
if plot_:
    _for_plot = abs(fft_now)
    plt.imshow(_for_plot.T, origin = "lower")
    plt.show()
    _for_plot = fft_now.real**2
    plt.imshow(_for_plot.T, origin = "lower")
    plt.show()
    _for_plot = fft_now.imag**2
    plt.imshow(_for_plot.T, origin = "lower")
    plt.show()

## Shifting origin
shift_ifft2 = np.fft.ifftshift(fft_now)

# IFFT
ifft = np.fft.ifft2(shift_ifft2)

## PLOT
if plot_:
    plt.imshow(ifft.real.T, origin = "lower")
    plt.show()
print(np.max(ifft.real.T), np.max(ifft.imag.T)) ## = np.max(images)

```

```python
def obs_make(images, obs_num, width):
    fft_now = np.fft.fft2(images)
    fft_now = np.fft.fftshift(fft_now)
    vis = np.zeros(np.shape(fft_now), dtype=np.complex)
    obs = 1000
    x_len, y_len = np.shape(images)
    position_vis = [[i,j] for i in range(int(x_len/2) - width,int(x_len/2) + width) for j in range(int(y_len/2) - width,int(y_len/2) + width)]
    
    Space_Position = np.array(position_vis).reshape(-1, 2)      # make it 2D
    random_indices = np.arange(0, Space_Position.shape[0])    # array of all indices
    np.random.shuffle(random_indices)                         # shuffle the array
    obs_vis_index = Space_Position[random_indices[:obs_num]]
    for i in range(obs_num):
        vix_x, vix_y = obs_vis_index[i][0], obs_vis_index[i][1]
        vis[vix_x, vix_y] = fft_now[vix_x, vix_y]
    return vis

obs_num = 200
vis_obs = obs_make(images, obs_num, 30)

_for_plot = abs(vis_obs)
plt.imshow(_for_plot.T, origin = "lower")
plt.show()
```

```python
vis_obs_shift = np.fft.ifftshift(vis_obs)

# IFFT
ifft_obs = np.fft.ifft2(vis_obs_shift)
print(np.max(ifft_obs), np.max(images))



fig, axs = plt.subplots(1, 2, figsize=(10, 3))

if plot_:
    im = axs[0].imshow(ifft_obs.real.T, origin = "lower")
    axs[0].set_title("zero padding")
    width_im = 10
    axs[0].set_xlim(int(x_len/2) - width_im,int(x_len/2) + width_im)
    axs[0].set_ylim(int(y_len/2) - width_im,int(y_len/2) + width_im)
    fig.colorbar(im)
    sum_flux_obs = np.sum(ifft_obs.real.T[int(x_len/2) - width_im:int(x_len/2) + width_im, int(y_len/2) - width_im:int(y_len/2) + width_im])
    
    im = axs[1].imshow(images.real.T, origin = "lower")
    width_im = 10
    axs[1].set_title("input images")
    axs[1].set_xlim(int(x_len/2) - width_im,int(x_len/2) + width_im)
    axs[1].set_ylim(int(y_len/2) - width_im,int(y_len/2) + width_im)
    fig.colorbar(im)
    sum_flux_model = np.sum(images.real.T[int(x_len/2) - width_im:int(x_len/2) + width_im, int(y_len/2) - width_im:int(y_len/2) + width_im])

plt.show()
normalization = sum_flux_model/sum_flux_obs
print(sum_flux_obs, sum_flux_model)
```

```python
import numpy as np
import numpy as np
import torch
import torch.nn as nn

def loss_function(model, *args):
    (obs, model_prior, lambda_l2) = args

    model_vis = np.fft.fftshift(np.fft.fft2(model))
    obs_mask = (obs == 0)
    d_vis = model_vis- obs
    d_vis[obs_mask] = 0
    return np.sum(np.abs(d_vis)**2) +  lambda_l2 * np.sum((model-model_prior)**2)
               
def loss_function_arr(model, *args):
    (obs, model_prior, lambda_l2) = args
    model_vis = np.fft.fftshift(np.fft.fft2(model))
    obs_mask = (obs == 0)
    d_vis = model_vis- obs
    d_vis[obs_mask] = 0
    return np.sum(np.abs(d_vis)**2), lambda_l2 * np.sum((model-model_prior)**2)


def grad_loss_arr(model, *args):
    
    (obs, model_prior, lambda_l2) = args

    ## Gradient for L2
    dl2_dmodel =2 * lambda_l2 * (model-model_prior)
    
    
    ## Gradient for chi^2
    model_vis = np.fft.fftshift(np.fft.fft2(model))
    obs_mask = (obs == 0)
    d_vis = model_vis - obs
    d_vis[obs_mask] = 0
    vis_F = np.fft.ifftshift(d_vis)
    ifft_F = np.fft.ifft2(d_vis)
    dF_dmodel =2* ifft_F.real 
    return dF_dmodel, dl2_dmodel 

def grad_loss(model, *args):
    
    (obs, model_prior, lambda_l2) = args

    ## Gradient for L2
    dl2_dmodel =2 * lambda_l2 * (model-model_prior)
    
    
    ## Gradient for chi^2
    model_vis = np.fft.fftshift(np.fft.fft2(model))
    obs_mask = (obs == 0)
    d_vis = model_vis - obs
    d_vis[obs_mask] = 0
    vis_F = np.fft.ifftshift(d_vis)
    ifft_F = np.fft.ifft2(d_vis)
    dF_dmodel =2* ifft_F.real 
    return dF_dmodel + dl2_dmodel 
    

def steepest_method( init_model, init_alpha, loss_function, graidient_function, \
                    ryo = 0.7,  c1=0.7, eps_stop = 1e-10, iter_max = 1000, *args):
    
    model = init_model
    alpha = init_alpha
    
    iter_num = 0
    
    while(1):
        gradient = graidient_function(model, *args)
        loss_1, loss_2 = loss_function_arr(model, *args)
        grad_1, grad_2 = grad_loss_arr(model, *args)

        print(iter_num, np.linalg.norm(grad_1), np.linalg.norm(grad_2),alpha)
        print(iter_num, loss_1, loss_2,alpha)

        
        if np.linalg.norm(gradient) < eps_stop:
            print("Iteration_stop")
            break

        while (1):
            model_new = model - alpha * gradient/np.linalg.norm(gradient)
            #print(alpha)
            if loss_function(model_new, *args) < loss_function(model, *args) - c1 *alpha* np.linalg.norm(gradient):
                break
            else:
                alpha = alpha * ryo
                
        model_prev =model
        model = model_new
        
        iter_num+=1
        #print(iter_num, np.sqrt(np.mean( (model - model_prev)**2))/np.sqrt(np.mean( (model)**2)))
        #print(iter_num, loss_function_arr(model_new, *args))

        if np.sqrt(np.mean( (model - model_prev)**2))/np.sqrt(np.mean( (model)**2))< eps_stop:
            break
        if iter_max < iter_num:
            break
    return model

obs = ifft_obs.real*normalization
gradient = grad_loss(obs, vis_obs, np.zeros(np.shape(ifft_obs.real/2)),1)
alpha = 1000 *(np.max(obs)/np.min(np.abs(gradient) ))
print(alpha)
```
