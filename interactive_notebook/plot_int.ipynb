```python
%matplotlib notebook
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np

file_name = "../../project_512_HD_3freqs/sparse_result/mfreq_solution.npz"
data = np.load(file_name)
print(data.files)
arr_sp_mfreq = data["image"]
fig, ax = plt.subplots()
im = ax.imshow(arr_sp_mfreq, origin = "lower")
fig.colorbar(im)


@interact
def show():
    im.set_array(arr_sp_mfreq)
    im.set_clim(arr_sp_mfreq.min(), arr_sp_mfreq.max())
    fig.canvas.draw_idle()

arr_sp_mfreq_1d = np.ravel(arr_sp_mfreq)
```

```python
file_name = "../../project_512_HD_3freqs/for_input/input_model.npz"
data = np.load(file_name)
print(data.files)
arr_input = data["model_nu0"]
fig, ax = plt.subplots()
im = ax.imshow(arr_input, origin = "lower")
fig.colorbar(im)

@interact
def show():
    im.set_array(arr_input)
    im.set_clim(arr_input.min(), arr_input.max())
    fig.canvas.draw_idle()

arr_input_1d = np.ravel(arr_input)
```
