```python
import nufft

nufft.nufft.nufft1d1freqs(10, 1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_range = np.linspace(-2,2,100)
eta  = 10
gauss = np.exp( - eta * x_range**2)
plt.plot(x_range, gauss)
```

```python
ms =100
df = 0.25
test = nufft.nufft.nufft1d1(x_range, gauss, ms, df=df , eps=1e-15, iflag=0, direct=False)
freq = nufft.nufft.nufft1d1freqs(ms, df=df )
test = nufft.nufft.nufft1d1(freq, test, ms, df=0.04 , eps=1e-15, iflag=1, direct=False)

```

```python
plt.plot(x_range, test.real*16)
plt.plot(x_range, gauss)
print(np.max(gauss)/ np.max(test.real)/np.pi)
```

```python
plt.plot(freq, test.real)
plt.plot(freq, 1/ ((2 * eta)**0.5) * np.exp(-(freq**2)/(4 * eta)) )
```

```python
test2 = np.mean(test.real/(1/ ((2 * eta)**0.5) * np.exp(-(freq**2)/(4 * eta))))
print(test2 * (np.pi )**0.5)
print(np.pi * 2 * 0.04)
```
